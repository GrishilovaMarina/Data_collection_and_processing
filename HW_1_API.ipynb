#2. Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search
###(а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа.
###(б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

```python
params = {
    'latitude': '55.80186',
    'longitude': '49.260556',
    'query' :  'манты'  
     }

url = 'https://api.delivery-club.ru/api1.2/vendors/search'

R = requests.get(url=url, params=params)

pprint(R.json())
```

```python
from statistics import mean

free_del = 0
pay_del = 0
min_order_free = []
min_order_pay = []
cat1_free = 0
cat5_free = 0
cat1_pay = 0
cat5_pay = 0
for i in R.json()['vendors']:
  if i['delivery']['price']['value'] == 0:
    free_del += 1
    min_order_free.append(i['delivery']['minOrderPrice']['value'])
    if i['categoryId'] == 1:
      cat1_free += 1
    else:
       cat5_free += 1
  else:
    pay_del += 1
    min_order_pay.append(i['delivery']['minOrderPrice']['value'])
    if i['categoryId'] == 1:
      cat1_pay += 1
    else:
       cat5_pay += 1
print(free_del, pay_del)
print(min_order_free)
print(min_order_pay)
print(f'Из предложенных точек доля с бесплатной доставкой составляет  {free_del/(free_del + pay_del):0%}, c платной - {pay_del/(free_del + pay_del):0%}')
print(f'Cредняя минимальная стоимость заказа при бесплатной доставке составляет   {mean(min_order_free)}, при платной - {mean(min_order_pay)}')
print(f'При бесплатной доставке доля магазинов составляет {cat5_free/(cat5_free + cat1_free):0%}, ресторанов - {cat1_free/(cat5_free + cat1_free):0%}')
print(f'При платной доставке доля магазинов составляет {cat5_pay/(cat5_pay + cat1_pay):0%}, ресторанов - {cat1_pay/(cat5_pay + cat1_pay):0%}')
```
