#Написать приложение или функцию, которые собирают основные новости с сайта на выбор lenta.ru, yandex-новости. Для парсинга использовать XPath.

```python
from pprint import pprint
from lxml import html
import requests
```

```python
url = 'https://lenta.ru/'

headers = {
    'User-Agent':	'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36',
}

R = requests.get(url = url, headers=headers)
```

```python
dom = html.fromstring(R.text)
list_main_news = dom.xpath("//section/div/div/a[@class='card-mini _compact']")
print(*list_main_news, sep = '\n')
```

```python
dict_main_news = {}

for i in list_main_news:
  news_source = dom.xpath("//head/title/text()")[0]
  news_name = i.xpath(".//span/text()")[0]
  news_link = url + i.xpath(".//@href")[0] 
  news_date = (i.xpath(".//@href")[0])[6:16] 
  dict_main_news[news_name] = {
      'источник' : news_source,
      'ссылка': news_link,
      'дата': news_date,
  }
for key in dict_main_news.keys():
  print(f'{key} - {dict_main_news[key]}')  
```

```python
url_2 = 'https://yandex.ru/news/'

R_2 = requests.get(url = url_2, headers = headers)
print(R_2.text)
```

```python
dom_2 = html.fromstring(R_2.text)
list_main_news_2 = dom_2.xpath("//section[1]/div/div")
print(list_main_news_2)
```

```python
dict_main_news_2 = {}
for n in list_main_news_2:
  news_source_2 = n.xpath(".//span/a/text()")[0]
  news_name_2 = n.xpath(".//h2/a/text()")[0]
  news_link_2 = n.xpath(".//span/a/@href")[0] 
  news_time_2 = n.xpath(".//span[@class='mg-card-source__time']/text()")[0]
  dict_main_news_2[news_name_2] = {
      'источник' : news_source_2,
      'ссылка': news_link_2,
      'время' : news_time_2
      }
for key in dict_main_news_2.keys():
  print(f'{key} - {dict_main_news_2[key]}') 
```
